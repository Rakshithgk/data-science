```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ufc_data= pd.read_csv('preprocessed_data.csv')
```

```python
ufc_data.head()
```

```python
ufc_data.shape
```

```python
ufc_data.isna().sum()
```

```python
ufc_data.dtypes
```

```python
Title_bout= pd.get_dummies(ufc_data['title_bout'], drop_first= True)
Title_bout.head()
```

```python
ufc_data['Title_bout']= Title_bout
```

```python
ufc_data.head()
```

```python
ufc_data.drop(['title_bout'], axis=1, inplace=True)
ufc_data.head()
```

```python
x= ufc_data.drop('Winner', axis= 1)
y= ufc_data['Winner']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix 
from sklearn.preprocessing import StandardScaler
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
scaler= StandardScaler()
x_std= scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test= train_test_split(x_std, y, test_size=0.3, random_state=1)
```

```python
model= LogisticRegression(random_state=0, multi_class='multinomial', solver= 'newton-cg')
model.fit(x_train, y_train)
```

```python
from sklearn import metrics
```

```python
y_predict= model.predict(x_test)
print(classification_report(y_test, y_predict))
print(confusion_matrix(y_test, y_predict))
print('accuracy:', metrics.accuracy_score(y_test, y_predict))
```

# Decision_tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model1= DecisionTreeClassifier(max_depth=4)
model1.fit(x_train, y_train)
```

```python
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus

dot_data= StringIO()
export_graphviz(model1,
               out_file=dot_data,
               filled=True,
               rounded=True,
               special_characters=True,
               class_names=['0','1'])

graph= pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('ufc.png')
Image(graph.create_png())
```

```python
y_predict= model1.predict(x_test)
print(classification_report(y_test, y_predict))
print(confusion_matrix(y_test, y_predict))
print('accuracy:', metrics.accuracy_score(y_test, y_predict))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model2= RandomForestClassifier(n_estimators=250)
model2.fit(x_train, y_train)

y_predict= model2.predict(x_test)
print(classification_report(y_test, y_predict))
print(confusion_matrix(y_test, y_predict))
print('accuracy:', metrics.accuracy_score(y_test, y_predict))
```

```python
from sklearn.svm import SVC
model3= SVC(kernel='linear', 
            class_weight='balanced',
            probability=True,
            gamma='auto')
model3.fit(x_train, y_train)
```

```python
y_predict= model3.predict(x_test)
print(classification_report(y_test, y_predict))
print(confusion_matrix(y_test, y_predict))
print('accuracy:', metrics.accuracy_score(y_test, y_predict))
```

```python
from sklearn.utils import resample
```

```python
# Separate majority and minority classes
ufc_data_majority = ufc_data[ufc_data.Winner=='Red']
ufc_data_minority = ufc_data[ufc_data.Winner=='Blue']
 
# Upsample minority class
ufc_data_minority_upsampled = resample(ufc_data_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=2380,    # to match majority class
                                 random_state=123) # reproducible results
 
# Combine majority class with upsampled minority class
ufc_data = pd.concat([ufc_data_majority, ufc_data_minority_upsampled])
 
# Display new class counts
ufc_data.Winner.value_counts()
```

```python
x= ufc_data.drop('Winner', axis= 1)
y= ufc_data['Winner']
```

```python
scaler= StandardScaler()
x_std= scaler.fit_transform(x)



x_train, x_test, y_train, y_test= train_test_split(x_std, y, test_size=0.3, random_state=1)

model= LogisticRegression(random_state=0, multi_class='multinomial', solver= 'newton-cg')
model.fit(x_train, y_train)

from sklearn import metrics

y_predict= model.predict(x_test)
print(classification_report(y_test, y_predict))
print(confusion_matrix(y_test, y_predict))
print('accuracy:', metrics.accuracy_score(y_test, y_predict))
```

```python
model2= RandomForestClassifier(n_estimators=250)
model2.fit(x_train, y_train)

y_predict= model2.predict(x_test)
print(classification_report(y_test, y_predict))
print(confusion_matrix(y_test, y_predict))
print('accuracy:', metrics.accuracy_score(y_test, y_predict))
```
