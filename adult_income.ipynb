```python
census_df.replace('?', np.nan, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
census_df.dtypes
```

```python
prepo= LabelEncoder()
census_df['workclass.ec']= prepo.fit_transform(census_df['workclass'])
census_df['marital.status.ec']= prepo.fit_transform(census_df['marital.status'])
census_df['occupation.ec']= prepo.fit_transform(census_df['occupation'])
census_df['relationship.ec']= prepo.fit_transform(census_df['relationship'])
census_df['race.ec']= prepo.fit_transform(census_df['race'])
census_df['sex.ec']= prepo.fit_transform(census_df['sex'])
census_df['native.country.ec']= prepo.fit_transform(census_df['native.country'])
```

```python
census_df.drop(['workclass', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country'], axis=1, inplace=True)
census_df.head()
```

```python
census_df['income.bi']= pd.get_dummies(census_df['income'], drop_first=True)
census_df.drop(['income'], axis=1, inplace=True)
census_df.head()
```

```python
sns.countplot(x= 'income.bi', data= census_df)
```

```python
census_df.drop(['fnlwgt', 'education'], axis=1, inplace=True)
```

```python
census_df.drop(['capital.gain','capital.loss'], axis=1, inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
x= census_df.drop(['income.bi'], axis=1)
y= census_df['income.bi']
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state= 1) 
```

```python
def analysis(x):
    model= x
    model.fit(x_train, y_train)
    y_pred= model.predict(x_test)
    
    print('error:', 1-metrics.accuracy_score(y_pred, y_test))
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
```

```python
analysis(x= LogisticRegression())
```

```python
analysis(x= GaussianNB())
```

```python
analysis(x= DecisionTreeClassifier(max_depth= 7))
```

```python
analysis(x= RandomForestClassifier(n_estimators= 250))
```
