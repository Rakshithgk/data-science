```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/omarish/5687264/raw/7e5c814ce6ef33e25d5259c1fe79463c190800d9/mpg.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df= df.replace('?', np.nan)
df= df.dropna()
```

```python
print(df.mpg.shape)
print(df.horsepower.shape)
print(df.weight.shape)
print(df.cylinders.shape)
```

```python
df.shape
```

```python
df= df.drop(['name', 'origin', 'model_year'], axis= 1)
x= df.drop('mpg', axis= 1)
y= df[['mpg']]
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 1)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
reg= LinearRegression()
reg.fit(x_train[['cylinders', 'horsepower', 'weight']], y_train)
y_predicted = reg.predict(x_test[['cylinders', 'horsepower', 'weight']])
print('mse: %.2f',  mean_squared_error(y_test, y_predicted))
print('r2: %.2f',  r2_score(y_test, y_predicted))
```
